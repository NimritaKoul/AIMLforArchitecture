# Optimization Algorithms

Optimization algorithms are computational methods designed to find the best possible solution to a problem by either minimizing costs/errors or maximizing profits/efficiency.

Regardless of the field, they work in an iterative loop:

- evaluating a current solution, measuring its performance, and making mathematical adjustments to get closer to the ideal outcome.

Every optimization problem and algorithm is built around three core components:

- Decision Variables: The adjustable inputs or parameters.
- The Objective Function: The exact mathematical formula that scores how "good" the decision variables are. This is the metric the algorithm wants to minimize (e.g., error, cost) or maximize (e.g., revenue, speed).
- Constraints: The physical or logical rules and limits that the solution must respect.

Understanding how these algorithms actually find the "best" answer comes down to their specific logic and approach.

Optimization algorithms are highly critical across a broad range of quantitative industries—serving as the engine that updates internal parameters to train artificial intelligence models, calculates optimal distribution routes in supply chain management, and maximizes structural integrity in manufacturing.

## The Core Iterative Loop

No matter how complex the problem, the basic logic of any optimization algorithm follows these steps:

- Initialization: Start with a random or baseline guess for the decision variables.

- Evaluation: Plug the variables into the objective function to get a score.

- Adjustment: Determine the direction and magnitude to change the variables to get a better score (this is where different algorithms diverge).

- Termination: Repeat steps 2 and 3 until the algorithm reaches a point where the score stops improving significantly (converges) or a time limit is reached.

Optimization methods can be broadly categorized based on how they navigate the search space and whether they utilize derivatives.

1. Derivative-Based vs. Derivative-Free

- Zero-Order Methods: Also called derivative-free or black-box algorithms, these rely entirely on evaluating the objective function's value directly. They are ideal when a function is noisy, non-differentiable, or completely hidden.

- First-Order Methods: These algorithms use the first derivative (gradient) of the objective function to map out the direction of steepest ascent or descent. They are computationally efficient and scale exceptionally well to large data.

- Second-Order Methods: These utilize both the gradient and the Hessian matrix (second derivative) to gauge the curvature of the function. They offer much faster convergence but demand significantly higher computational memory and power.

2. Deterministic vs. Stochastic

- Deterministic Algorithms: These follow strict, predefined mathematical rules. Initializing the algorithm with the exact same inputs will always produce the identical output path. They work best for well-structured or convex problems.

- Stochastic Algorithms: These inject randomness into the search process. By using probabilistic choices, they avoid getting stuck in local bottlenecks (local minima) and are better suited to explore complex, non-convex landscapes.


## Key Logic Paradigms In Optimization Algorithms

1. Gradient-Based (First and Second Order): These algorithms use the mathematical slope (derivative) of the objective function to find the optimal point.

The Logic: If you want to find the lowest point in a valley, you feel the slope and walk downhill.

Usage: They are the primary engine behind Machine Learning and training neural networks.

Examples: Gradient Descent, Stochastic Gradient Descent (SGD)

2. Metaheuristic & Evolutionary Algorithms

These algorithms rely on randomness and heuristics rather than precise math or derivatives. They are designed for "black-box" problems where standard math is too difficult or the function has too many variables.

The Logic: Mimic natural phenomena like biological evolution, metal cooling (annealing), or animal swarms. They maintain an entire "population" of guesses, constantly sharing the best traits, mutating them, and weeding out the worst.

Usage: Logistics routing, assembly line balancing, and complex design trade-offs.

Examples: Genetic Algorithms, Particle Swarm Optimization (PSO), and Simulated Annealing

3. Linear and Integer Programming (Deterministic)

When problems have strict rules (like scheduling or resource allocation), these deterministic algorithms systematically search for the absolute best mathematical solution.

The Logic: Instead of guessing, they represent the constraints as a multi-dimensional geometric space and methodically traverse the edges or boundaries to find the highest/lowest point.

Usage: Supply chain management, Portfolio Optimization, and manufacturing

## Applying optimization to architectural planning

Applying optimization to architectural planning, often implemented via Generative Design or Architectural Design Optimization (ADO),inverts the traditional design workflow. Instead of sketching a shape and simulating its performance afterward, you feed your constraints and goals into an algorithm to co-create the ideal structure.

The exact algorithm you choose depends on which part of the planning process you are trying to solve:

1. Floor Plan Layout & Spatial Topology

Determining where rooms should go, how large they should be, and how they connect to one another is a complex, combinatorial problem.

- The Problem: Placing rooms to maximize convenient foot-traffic flow, maintain natural groupings (e.g., kitchen near dining room), and fit within a specific total footprint.

- The Logic: This is an NP-hard problem split into topology (logical relationships) and geometry (sizes and coordinates).

- Best Algorithms: Genetic Algorithms (GAs) or Multi-Population Genetic Algorithms (MPGA). These simulate variations of room placements as "chromosomes," crossing them over and mutating them across generations to evolve the most efficient layout.

- Industry Tools: Autodesk Generative Design (within Revit) or Grasshopper scripts paired with the Galapagos optimizer.

2. Building Massing & Environmental Performance

Optimizing the external shape, orientation, and window placement of a building to react to its surrounding environment.

- The Problem: Shifting the building's shape to maximize natural daylight (reducing electricity bills), maximize solar panel exposure, and minimize intense summer heat.

- The Logic: This is a Multi-Objective Optimization (MOO) problem. Because goals often conflict (e.g., more windows means more natural light but also more heat loss in winter), the algorithm cannot give you one single answer. Instead, it generates a Pareto Frontier—a curated collection of optimal trade-offs.

- Best Algorithms: NSGA-II (Non-dominated Sorting Genetic Algorithm II) or Particle Swarm Optimization (PSO).

- Industry Tools: Autodesk Forma (great for microclimate and wind simulations), or Grasshopper paired with Wallacei or Octopus.

3. Structural & Material Efficiency (Topology Optimization)
Finding the most efficient structural form to hold a building up using the least amount of material.

- The Problem: Designing complex trusses, columns, or canopy roofs that look organic but are mathematically perfect for handling load stresses.

- The Logic: The algorithm treats a block of space as a fine grid of voxels. It applies simulated physical forces (gravity, wind) and systematically subtracts material from areas that experience no stress.

- Best Algorithms: Gradient-based optimization combined with Finite Element Method (FEM) analysis. Because physics follows predictable mathematical curves, derivatives work flawlessly here to minimize material volume while maximizing stiffness.

- Industry Tools: Ameba (for Grasshopper) or Altair Inspire.

# Optimization Algorithms in Machine Learning

Machine learning models learn by minimizing a loss function that measures the difference between predicted and actual values.

Optimization algorithms are used to update model parameters so that this loss is reduced and the model learns better from data.

Their main roles in training include:

- Adjust model parameters to minimize the loss function.
- Help models improve predictions through repeated updates.
- Move the model toward an optimal solution during training.

## First Order Algorithms
First order optimization algorithms use the first derivative (gradient) of the loss function to update model parameters and move toward an optimal solution. They are widely used in machine learning because they are computationally efficient and scale well to large datasets.

### Gradient Descent
Gradient Descent is a first order optimization algorithm used to minimize a loss function by updating model parameters step by step in the direction that reduces the error. It is widely used in machine learning and deep learning to train models efficiently.

- Compute gradient: Calculate the derivative of the loss function at the current parameters.
- Update parameters: Move parameters in the opposite direction of the gradient using a learning rate.
- Repeat process: Continue updating parameters for multiple iterations.
Reach minimum: Parameters gradually move toward values that minimize the loss.

#### Variants of Gradient Descent

Batch Gradient Descent: Uses the entire dataset to compute the gradient and update parameters. It is stable but can be slow for very large datasets.
Stochastic Gradient Descent (SGD): Updates model parameters using one training example at a time. It requires less computation and works well for large datasets.
Mini Batch Gradient Descent: Updates parameters using small batches of data instead of the entire dataset. It balances speed and stability and is widely used to train deep learning models.

#### Implementation

- Defines the function and its gradient: Specifies the function and how its slope is calculated.
- Applies Gradient Descent: Starts from an initial value and updates it repeatedly using the learning rate.
- Checks convergence: Stops when the change in the value becomes very small.
- Visualizes the optimization: Plots the function and shows how Gradient Descent moves toward the minimum point.

```python
import numpy as np
import matplotlib.pyplot as plt
def function(x):
    return x**2
def gradient(x):
    return 2 * x
def gradient_descent(start, learning_rate=0.1, n_iter=50, tolerance=1e-6):
    x = start
    history = [x]

    for i in range(n_iter):
        grad = gradient(x)
        new_x = x - learning_rate * grad

        if abs(new_x - x) < tolerance:
            print(f"Converged at iteration {i+1}")
            break

        x = new_x
        history.append(x)

    return x, history

start = 5.0
learning_rate = 0.1
n_iter = 50
tolerance = 1e-6
result, history = gradient_descent(start, learning_rate, n_iter, tolerance)
print("Final optimized value of x:", result)
print("Minimum value of f(x):", function(result))
x_vals = np.linspace(-6, 6, 400)
y_vals = function(x_vals)
plt.figure()
plt.plot(x_vals, y_vals)
plt.scatter(history, [function(x) for x in history])
plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x) = x^2")
plt.show()
```


## Stochastic Optimization Techniques
Stochastic optimization techniques use randomness in the optimization process to explore the search space more effectively. They are useful for solving complex problems where traditional optimization methods may get stuck in local minima.

- Simulated Annealing: It is an optimization algorithm inspired by metallurgy. It escapes local minima by gradually reducing randomness to find near optimal solutions.
- Random Search: A simple method that randomly selects points in the search space and evaluates them. It is surprisingly effective for high dimensional optimization problems.

## Evolutionary Algorithms
Evolutionary Algorithms are optimization methods inspired by natural selection. They work by evolving a group of candidate solutions over multiple generations to find good or near optimal solutions for complex problems.

Main components of Evolutionary Algorithms are:

- Population: A group of possible solutions.
- Fitness Function: Measures how good each solution is.
- Selection: Chooses the better solutions for reproduction.
- Genetic Operators: Create new solutions using crossover and mutation.
- Termination: Stops the process after a condition is met.

Techniques in Evolutionary Algorithms:

### Genetic Algorithms

Genetic Algorithms are optimization methods that evolve a population of solutions using selection, crossover and mutation.

- Start with randomly generated solutions
- Select the best solutions based on fitness
- Create new solutions through crossover and mutation
- Repeat the process for several generations

### Differential Evolution (DE)

Differential Evolution is a population based stochastic optimization technique that evolves candidate solutions using vector differences. It is effective for continuous, high dimensional and non convex optimization problems.

- Starts with a population of candidate solutions within defined bounds.
- For each target vector, generates a mutant vector using three distinct population members with the formula:

mutant = a +F .(b -c), where F is a scaling factor controlling variation

- Creates a trial vector by mixing the target and mutant vectors based on a crossover rate (CR).
- Replaces the target vector with the trial vector if it improves fitness and repeats the process for a fixed number of generations.


## Metaheuristic Optimization
Metaheuristic optimization algorithms are high level methods used to solve complex optimization problems by exploring large search spaces. They help find near optimal solutions without using gradient information.

### Common Techniques

1. Tabu Search (TS)

Tabu Search improves local search by using memory to avoid revisiting recently explored solutions and helps escape local optima.

- Uses a tabu list to avoid repeating recent solutions.
- Allows better solutions using aspiration rules.
= Explores new areas while improving good solutions.

2. Iterated Local Search (ILS)

Iterated Local Search is another strategy for enhancing local search but unlike Tabu Search, it does not use memory structures. It relies on repeated application of local search, combined with random changes to escape local minima and continue the search.

- Starts with a local search to find a good solution.
- Applies small changes to escape local minima.
- Repeats the process to improve the solution.

5. Swarm Intelligence Algorithms

Swarm intelligence algorithms are inspired by the collective behaviour of natural systems such as bird flocks and ant colonies. These algorithms use multiple agents that cooperate and share information to search for good solutions.

There are two of the widely applied algorithms in swarm intelligence:

5.1. Particle Swarm Optimization (PSO)

Particle Swarm Optimization (PSO) is a population based algorithm inspired by the social behaviour of birds and fish. Each particle represents a potential solution and moves through the search space by combining its own experience with that of the swarm.

- Particles update their velocity based on personal best and global best solutions.
- Their position is updated using the new velocity.
- Over iterations, particles move toward better solutions.

5.2. Ant Colony Optimization (ACO)

Ant Colony Optimization ACO is inspired by ant foraging behaviour. Ants find shortest paths by depositing pheromones, which guide other ants toward better solutions over iterations.

- Each ant builds a path by choosing options based on pheromone strength.
- Pheromones evaporate over time and are reinforced on better paths.
- Over iterations, ants converge toward the shortest path.

6. Optimization Techniques in Deep Learning

Deep learning models often contain many parameters, making optimization important for efficient training. Different optimization techniques help models learn faster and improve prediction performance.

Common optimizers used in Neural Networks are:

- RMSProp: Adjusts the learning rate for each parameter using a moving average of squared gradients, which helps stabilize training and improves convergence.
- Adam: Combines ideas from momentum and RMSProp by maintaining moving averages of gradients and their squares. It adapts learning rates for each parameter and is widely used for training deep learning models.

7. Hyperparameter Optimization

Hyperparameter optimization is the process of selecting the best hyperparameter values to improve a machine learning model's performance. These parameters are not learned from data but strongly affect accuracy, efficiency and generalization.

- Grid Search: Tests all possible combinations of predefined hyperparameter values to find the best one, but it can be computationally expensive.
- Random Search: Randomly selects hyperparameter values from a defined range. It is faster and often works well for large or high dimensional search spaces.

## Second order Optimization Algorithms

Second order optimization algorithms use both the gradient and second derivative of the loss function to update parameters more accurately. They often converge faster than first order methods but are computationally more expensive.

1. Newton Method
Newton's Method is an optimization technique that uses both the gradient and second derivative of a function to update parameters more accurately and reach the minimum faster than basic gradient based methods.

- Calculate the gradient (first derivative) of the function at the current point.
- Compute the Hessian matrix (second derivative) to understand the curvature.
Update parameters using:

θnew =θold-Hexp-1⋅∇f(θold)

- Repeat the process until the change in parameters is very small or convergence is reached.

#### Implementation of Newton Method

- Defines the function and its derivatives: Specifies the function along with its first and second derivatives.
- Applies Newton's Method: Starts from an initial value and repeatedly updates it using gradient and curvature information.
- Checks convergence: Stops when the update becomes very small or when the maximum number of iterations is reached.
- Visualizes the result: Plots the function and highlights the stationary point found by the algorithm.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 2*x**2 + 2

def f_prime(x):
    return 3*x**2 - 4*x

def f_double_prime(x):
    return 6*x - 4

def newtons_method(f_prime, f_double_prime, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        second_derivative = f_double_prime(x)
        if second_derivative == 0:
            print("Zero second derivative. Stopping at iteration", i)
            break
        step = f_prime(x) / second_derivative
        if abs(step) < tol:
            print(f"Converged after {i+1} iterations")
            break
        x -= step
    return x, f(x)

x0 = 3.0
tol = 1e-6
max_iter = 100
result, f_val = newtons_method(f_prime, f_double_prime, x0, tol, max_iter)
print("Stationary point at x =", result)
print("Function value at this point f(x) =", f_val)
x_vals = np.linspace(-1, 4, 500)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label="f(x) = x³ - 2x² + 2", color='blue')
plt.scatter(result, f_val, color='red', zorder=5, label="Stationary Point")
plt.title("Newton's Method Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

2. Quasi-Newton Methods

Quasi-Newton methods are optimization algorithms that find local minima using gradient information and an approximation of the function’s curvature. Instead of computing the Hessian matrix directly, they estimate it, making the optimization process faster and more efficient.

- Avoid full Hessian computation: Reduces computational cost.
- Approximate curvature: Uses gradient information to estimate second order behaviour.
- Faster convergence: Often converges faster than basic gradient descent.

Quasi-Newton Variants

- BFGS (Broyden-Fletcher-Goldfarb-Shanno): Estimates the Hessian matrix using gradient updates and improves it iteratively, achieving fast convergence without computing the Hessian directly.
- L-BFGS (Limited-memory BFGS): A memory efficient version of BFGS that stores only a few past updates, making it suitable for large scale optimization problems.

3. Constrained Optimization

Constrained optimization deals with optimizing an objective function while satisfying certain restrictions or constraints. These constraints can be equality or inequality conditions and specialized methods help find optimal solutions that respect them.

- Lagrange Multipliers: Introduce additional variables called multipliers to convert a constrained problem with equality constraints into an unconstrained one.

- KKT (Karush-Kuhn-Tucker) Conditions: Extend Lagrange multipliers to handle both equality and inequality constraints and provide conditions for optimal solutions.

4. Bayesian Optimization
Bayesian optimization is a probabilistic method used to optimize expensive or complex functions. It builds a surrogate model of the objective function and uses past evaluations to decide where to search next.

- Uses past evaluations: Guides the search using information from previously tested points.

- Sample efficient: Requires fewer function evaluations than methods like Grid Search or Random Search.

- Balances exploration and exploitation: Searches new areas while focusing on promising regions.

# Genetic Algorithm (GA) to optimize an architectural floor plan layout based on room adjacency preferences and boundaries.

In the below cell, we have implemented a Genetic Algorithm (GA) to optimize an architectural floor plan layout based on room adjacency preferences and boundaries. It arranges rooms on a grid to minimize distances between highly related zones (e.g., keeping the Kitchen near the Living Room).

## Core Logic & Components

- Chromosome Representation: Each layout is a dictionary, mapping rooms to discrete 2D grid coordinates (x, y).

- Fitness Function: Calculates the Euclidean distance between all room pairs. It prioritizes placing rooms closer together if they have high weights in the adjacency matrix.

- Selection: Uses roulette wheel selection based on normalized fitness scores.

- Crossover: Swaps room coordinates between two parent layouts.

- Mutation: Randomly shifts a room to an empty neighboring grid space.


```python
import random
import math

# --- 1. CONFIGURATION & ARCHITECTURAL DATA ---
GRID_WIDTH = 5
GRID_HEIGHT = 5
POPULATION_SIZE = 20
GENERATIONS = 50
MUTATION_RATE = 0.2

ROOMS = ['Living Room', 'Kitchen', 'Bedroom', 'Bathroom', 'Corridor']

# High values mean rooms should be close together (adjacency matrix)
ADJACENCY_MATRIX = {
    ('Living Room', 'Kitchen'): 5,
    ('Living Room', 'Corridor'): 4,
    ('Bedroom', 'Bathroom'): 5,
    ('Bedroom', 'Corridor'): 4,
    ('Bathroom', 'Corridor'): 2,
    ('Kitchen', 'Corridor'): 1,
}

def get_adjacency_weight(r1, r2):
    """Retrieves adjacency priority weight between two rooms."""
    return ADJACENCY_MATRIX.get((r1, r2), ADJACENCY_MATRIX.get((r2, r1), 0))

# --- 2. GENETIC ALGORITHM UTILITIES ---
def generate_random_layout():
    """Creates a random unique coordinate layout for all rooms."""
    all_positions = [(x, y) for x in range(GRID_WIDTH) for y in range(GRID_HEIGHT)]
    chosen_positions = random.sample(all_positions, len(ROOMS))
    return {room: pos for room, pos in zip(ROOMS, chosen_positions)}

def calculate_fitness(layout):
    """Calculates fitness. Closer highly-adjacent rooms produce higher scores."""
    total_score = 0
    for i in range(len(ROOMS)):
        for j in range(i + 1, len(ROOMS)):
            r1, r2 = ROOMS[i], ROOMS[j]
            weight = get_adjacency_weight(r1, r2)
            if weight > 0:
                p1, p2 = layout[r1], layout[r2]
                # Calculate Euclidean distance between room centers
                dist = math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                # Avoid division by zero; minimize distance by maximizing inverse
                total_score += weight / (dist + 0.1)
    return total_score

def select_parent(population, fitness_scores):
    """Selects a parent layout using Roulette Wheel Selection."""
    total_fitness = sum(fitness_scores)
    pick = random.uniform(0, total_fitness)
    current = 0
    for individual, score in zip(population, fitness_scores):
        current += score
        if current > pick:
            return individual
    return population[-1]

def crossover(parent1, parent2):
    """Combines two layouts while resolving duplicate grid slot collisions."""
    child = {}
    # Randomly split the room inheritance
    split = random.randint(1, len(ROOMS) - 1)

    # Inherit first half from parent1
    for room in ROOMS[:split]:
        child[room] = parent1[room]

    # Inherit remainder from parent2 if slots are open
    for room in ROOMS[split:]:
        pos = parent2[room]
        if pos not in child.values():
            child[room] = pos
        else:
            # Fallback: Find first available empty grid space
            all_positions = [(x, y) for x in range(GRID_WIDTH) for y in range(GRID_HEIGHT)]
            random.shuffle(all_positions)
            for alt_pos in all_positions:
                if alt_pos not in child.values():
                    child[room] = alt_pos
                    break
    return child

def mutate(layout):
    """Mutates a layout by changing a random room's position to an open space."""
    if random.random() < MUTATION_RATE:
        room_to_move = random.choice(ROOMS)
        all_positions = [(x, y) for x in range(GRID_WIDTH) for y in range(GRID_HEIGHT)]
        occupied = list(layout.values())
        empty_positions = [pos for pos in all_positions if pos not in occupied]

        if empty_positions:
            layout[room_to_move] = random.choice(empty_positions)
    return layout

def print_layout(layout):
    """Renders a scannable text grid representation of the floor plan."""
    grid = [["[        ]" for _ in range(GRID_WIDTH)] for _ in range(GRID_HEIGHT)]
    for room, (x, y) in layout.items():
        grid[y][x] = f"[{room[:8]:^8}]"
    for row in grid:
        print(" ".join(row))

# --- 3. MAIN EVOLUTIONARY LOOP ---
def run_genetic_algorithm():
    # Initialize Population
    population = [generate_random_layout() for _ in range(POPULATION_SIZE)]
    best_layout = None
    best_fitness = -1

    print(f"Starting Evolution over {GENERATIONS} generations...\n")

    for generation in range(GENERATIONS):
        fitness_scores = [calculate_fitness(ind) for ind in population]

        # Track historical best option
        for ind, score in zip(population, fitness_scores):
            if score > best_fitness:
                best_fitness = score
                best_layout = ind

        # Breed new generation
        next_population = []
        for _ in range(POPULATION_SIZE // 2):
            p1 = select_parent(population, fitness_scores)
            p2 = select_parent(population, fitness_scores)

            c1 = mutate(crossover(p1, p2))
            c2 = mutate(crossover(p2, p1))

            next_population.extend([c1, c2])

        population = next_population

    print(f"Evolution Completed. Optimized Layout Fitness Score: {best_fitness:.2f}\n")
    print_layout(best_layout)


run_genetic_algorithm()
```

To handle room dimensions (width × height) and window orientation constraints, we must shift from a simple coordinate point system to a 2D Box Packing / Boundary Representation model.

The below code expands above genetic algorithm and assigns each room an entry-point coordinate, a discrete width, and a discrete height. The fitness function penalizes overlapping rooms and rewards placing specific rooms (like the Living Room or Bedroom) along external walls with the correct solar/daylight window orientation.

Core Architectural Extensions

- Expanded Gene Footprint: Each room is represented by a sequence of 4 variables: (x, y, width, height).

- Overlap Penalty: Since rooms can now occupy multiple grid tiles, the script actively calculates overlapping areas and heavily docks the fitness score if spaces clip through each other.

- Daylight Orientation Constraints: A dedicated function checks if specific spaces sit on the perimeter of the building boundary corresponding to their required orientation (e.g., Living Room facing South for passive solar gain).

Key Performance Variables to Tweak

- Overlap Penalty Weights: The script multiplies room overlapping area by 250. If your generated layouts are still messy or overlapping, increase this parameter to 1000 to completely eliminate layout intersections.

- Site Dimensions: Adjusting SITE_WIDTH and SITE_HEIGHT alters the density constraint. Tight site constraints simulate small urban infill spaces, forcing rooms to pack densely.

```python
import random
import math

# --- 1. CONFIGURATION & ARCHITECTURAL DATA ---
# Define the outer bounding box of the building site layout
SITE_WIDTH = 12
SITE_HEIGHT = 12

POPULATION_SIZE = 30
GENERATIONS = 100
MUTATION_RATE = 0.3

ROOM_CONFIGS = {
    # 'RoomName': (min_w, max_w, min_h, max_h)
    'Living Room': (4, 6, 4, 6),
    'Kitchen':     (3, 4, 3, 5),
    'Bedroom':     (3, 5, 3, 5),
    'Bathroom':    (2, 3, 2, 4),
    'Corridor':    (2, 2, 4, 6)
}
ROOMS = list(ROOM_CONFIGS.keys())

ADJACENCY_MATRIX = {
    ('Living Room', 'Kitchen'): 5,
    ('Living Room', 'Corridor'): 4,
    ('Bedroom', 'Bathroom'): 5,
    ('Bedroom', 'Corridor'): 4,
    ('Bathroom', 'Corridor'): 2,
}

# Daylight rules: Preferred cardinal direction for window placement
DAYLIGHT_CONSTRAINTS = {
    'Living Room': 'South',
    'Bedroom': 'East',
    'Kitchen': 'North'
}

def get_adjacency_weight(r1, r2):
    return ADJACENCY_MATRIX.get((r1, r2), ADJACENCY_MATRIX.get((r2, r1), 0))

# --- 2. EVALUATION & CONSTRAINTS ENGINE ---
def calculate_overlap_area(rect1, rect2):
    """Calculates the shared intersection area between two room rectangles."""
    x1, y1, w1, h1 = rect1
    x2, y2, w2, h2 = rect2

    # Find intersecting bounds
    x_left = max(x1, x2)
    y_top = max(y1, y2)
    x_right = min(x1 + w1, x2 + w2)
    y_bottom = min(y1 + h1, y2 + h2)

    if x_right > x_left and y_bottom > y_top:
        return (x_right - x_left) * (y_bottom - y_top)
    return 0

def check_daylight_compliance(room_rect, orientation):
    """Checks if a room touches the perimeter wall of its preferred direction."""
    x, y, w, h = room_rect
    if orientation == 'North' and (y + h) >= SITE_HEIGHT: return True
    if orientation == 'South' and y <= 0: return True
    if orientation == 'East' and (x + w) >= SITE_WIDTH: return True
    if orientation == 'West' and x <= 0: return True
    return False

def calculate_fitness(layout):
    """Evaluates spatial layout based on adjacency, daylighting, and overlap limits."""
    adjacency_score = 0
    overlap_penalty = 0
    daylight_reward = 0
    out_of_bounds_penalty = 0

    # 1. Adjacency & Distance calculation between center points
    for i in range(len(ROOMS)):
        for j in range(i + 1, len(ROOMS)):
            r1, r2 = ROOMS[i], ROOMS[j]
            w1, h1 = layout[r1][2], layout[r1][3]
            w2, h2 = layout[r2][2], layout[r2][3]

            # Center points
            c1_x, c1_y = layout[r1][0] + w1/2, layout[r1][1] + h1/2
            c2_x, c2_y = layout[r2][0] + w2/2, layout[r2][1] + h2/2

            weight = get_adjacency_weight(r1, r2)
            if weight > 0:
                dist = math.sqrt((c1_x - c2_x)**2 + (c1_y - c2_y)**2)
                adjacency_score += weight / (dist + 0.1)

    # 2. Heavy Penalty for Room Overlaps & Boundaries
    for i, r1 in enumerate(ROOMS):
        x1, y1, w1, h1 = layout[r1]

        # Check Site Boundary Violations
        if x1 + w1 > SITE_WIDTH or y1 + h1 > SITE_HEIGHT or x1 < 0 or y1 < 0:
            out_of_bounds_penalty += 500

        for j, r2 in enumerate(ROOMS):
            if i != j:
                overlap = calculate_overlap_area(layout[r1], layout[r2])
                if overlap > 0:
                    overlap_penalty += overlap * 250

    # 3. Daylight Window Orientation Bonus
    for room, orientation in DAYLIGHT_CONSTRAINTS.items():
        if check_daylight_compliance(layout[room], orientation):
            daylight_reward += 50  # Positive reinforcement for good solar access

    # Combine components: Maximize adjacency/daylight, minimize overlaps/out-of-bounds
    total_fitness = adjacency_score + daylight_reward - overlap_penalty - out_of_bounds_penalty
    return max(0.01, total_fitness) # Prevent zero or negative fitness value drops

# --- 3. GENETIC OPERATORS ---
def generate_random_layout():
    """Generates a randomized footprint layout matching individual room size parameters."""
    layout = {}
    for room, bounds in ROOM_CONFIGS.items():
        min_w, max_w, min_h, max_h = bounds
        w = random.randint(min_w, max_w)
        h = random.randint(min_h, max_h)
        # Position within the site constraints safely
        x = random.randint(0, max(0, SITE_WIDTH - w))
        y = random.randint(0, max(0, SITE_HEIGHT - h))
        layout[room] = [x, y, w, h]
    return layout

def select_parent(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    pick = random.uniform(0, total_fitness)
    current = 0
    for individual, score in zip(population, fitness_scores):
        current += score
        if current > pick:
            return individual
    return population[-1]

def crossover(parent1, parent2):
    """Blends genetic sizes and position markers from parents."""
    child = {}
    split = random.randint(1, len(ROOMS) - 1)
    for i, room in enumerate(ROOMS):
        if i < split:
            child[room] = list(parent1[room])
        else:
            child[room] = list(parent2[room])
    return child

def mutate(layout):
    """Mutates either the position, or dimensions of random rooms."""
    if random.random() < MUTATION_RATE:
        room = random.choice(ROOMS)
        mutation_type = random.choice(['position', 'dimension'])
        min_w, max_w, min_h, max_h = ROOM_CONFIGS[room]

        if mutation_type == 'position':
            layout[room][0] = random.randint(0, max(0, SITE_WIDTH - layout[room][2]))
            layout[room][1] = random.randint(0, max(0, SITE_HEIGHT - layout[room][3]))
        else:
            layout[room][2] = random.randint(min_w, max_w)
            layout[room][3] = random.randint(min_h, max_h)
    return layout

def print_layout_matrix(layout):
    """Visualizes the 2D dimensional layout boundaries directly inside the terminal console."""
    # Build text representation grid matrix
    grid = [[" . " for _ in range(SITE_WIDTH)] for _ in range(SITE_HEIGHT)]

    # Render allocations from top-down perspective
    for room, (x, y, w, h) in layout.items():
        char = room[0] # Use first letter of room as visual tag
        for r_y in range(y, min(y + h, SITE_HEIGHT)):
            for r_x in range(x, min(x + w, SITE_WIDTH)):
                grid[r_y][r_x] = f" {char} "

    # Flip grid upside down so South corresponds to Y=0 indices visually
    for row in reversed(grid):
        print("".join(row))
    print(f"\nLegend: L=Living Room, K=Kitchen, B=Bedroom, b=Bathroom, C=Corridor")

# --- 4. EXECUTION PIPELINE ---
def run_advanced_ga():
    population = [generate_random_layout() for _ in range(POPULATION_SIZE)]
    best_layout = None
    best_fitness = -1

    print(f"Optimizing Dimensional Footprints & Daylight Orientation over {GENERATIONS} generations...\n")

    for generation in range(GENERATIONS):
        fitness_scores = [calculate_fitness(ind) for ind in population]

        for ind, score in zip(population, fitness_scores):
            if score > best_fitness:
                best_fitness = score
                best_layout = ind

        next_population = []
        for _ in range(POPULATION_SIZE // 2):
            p1 = select_parent(population, fitness_scores)
            p2 = select_parent(population, fitness_scores)

            c1 = mutate(crossover(p1, p2))
            c2 = mutate(crossover(p2, p1))

            next_population.extend([c1, c2])
        population = next_population

    print(f"Evolution Completed. Top Layout Fitness Score: {best_fitness:.2f}")
    print("\n--- OPTIMIZED FLOOR PLAN MAP (North ▲ / South ▼) ---")
    print_layout_matrix(best_layout)

    print("\nRoom Allocation Specifications:")
    for room, (x, y, w, h) in best_layout.items():
        print(f" - {room:<12}: Origin (X:{x}, Y:{y}) | Dimensions: {w}w x {h}h")


run_advanced_ga()
```


<b> In the code below, we introduce structural wall thickness and graph-based circulation path validation, and shift from simple geometry checks to a network topology approach.</b>

We will use an Adjacency Graph to ensure connectivity. A room is considered accessible only if it shares an unblocked intersecting boundary (a "doorway") with a corridor or an adjacent room that eventually connects to the main entrance.

Core Architectural Extensions

- Wall Thickness Offsets: Rooms are expanded by a set wall thickness parameter during overlap checks. This ensures structural partitions fit between rooms without clipping into usable floor area.

- Graph-Based Connectivity Check: The script treats rooms as nodes and shared wall segments as potential door links. It builds an adjacency graph and runs a Breadth-First Search (BFS) starting from the Corridor (acting as the circulation spine). Any room isolated from the network triggers a heavy fitness penalty

Verification & Rules

- share_accessible_wall(): Acts as a gateway builder. If two rooms are perfectly flushed edge-to-edge with at least 1 unit length intersection, a valid door link is registered.

- circulation_penalty: Set to 1000. This creates a severe adaptive constraint. Layout variations that trap rooms without door access are aggressively pruned from the breeding pool.

```python
import random
import math

# --- 1. CONFIGURATION & ARCHITECTURAL DATA ---
SITE_WIDTH = 14
SITE_HEIGHT = 14

POPULATION_SIZE = 40
GENERATIONS = 120
MUTATION_RATE = 0.35
WALL_THICKNESS = 1 # Structural envelope padding around rooms

ROOM_CONFIGS = {
    'Corridor':    (2, 2, 5, 7), # Act as the central circulation spine
    'Living Room': (4, 6, 4, 6),
    'Kitchen':     (3, 4, 3, 4),
    'Bedroom':     (3, 5, 3, 4),
    'Bathroom':    (2, 3, 2, 3)
}
ROOMS = list(ROOM_CONFIGS.keys())

ADJACENCY_MATRIX = {
    ('Living Room', 'Kitchen'): 5,
    ('Living Room', 'Corridor'): 5,
    ('Bedroom', 'Bathroom'): 4,
    ('Bedroom', 'Corridor'): 5,
    ('Bathroom', 'Corridor'): 3,
    ('Kitchen', 'Corridor'): 3,
}

DAYLIGHT_CONSTRAINTS = {
    'Living Room': 'South',
    'Bedroom': 'East',
    'Kitchen': 'North'
}

def get_adjacency_weight(r1, r2):
    return ADJACENCY_MATRIX.get((r1, r2), ADJACENCY_MATRIX.get((r2, r1), 0))

# --- 2. ADVANCED GEOMETRY & GRAPH VALIDATION ---
def calculate_overlap_with_walls(rect1, rect2):
    """Calculates overlaps including structural wall thickness buffer around zones."""
    x1, y1, w1, h1 = rect1
    x2, y2, w2, h2 = rect2

    # Expand boundaries by wall thickness for intersection testing
    x_left = max(x1 - WALL_THICKNESS, x2 - WALL_THICKNESS)
    y_top = max(y1 - WALL_THICKNESS, y2 - WALL_THICKNESS)
    x_right = min(x1 + w1 + WALL_THICKNESS, x2 + w2 + WALL_THICKNESS)
    y_bottom = min(y1 + h1 + WALL_THICKNESS, y2 + h2 + WALL_THICKNESS)

    if x_right > x_left and y_bottom > y_top:
        return (x_right - x_left) * (y_bottom - y_top)
    return 0

def share_accessible_wall(rect1, rect2):
    """Returns True if rooms share a wall segment long enough to host a door."""
    x1, y1, w1, h1 = rect1
    x2, y2, w2, h2 = rect2

    # Check horizontal adjacency (Left/Right touch)
    if (x1 + w1 == x2) or (x2 + w2 == x1):
        overlap_y = min(y1 + h1, y2 + h2) - max(y1, y2)
        if overlap_y >= 1: # At least 1 unit shared wall space for door aperture
            return True

    # Check vertical adjacency (Top/Bottom touch)
    if (y1 + h1 == y2) or (y2 + h2 == y1):
        overlap_x = min(x1 + w1, x2 + w2) - max(x1, x2)
        if overlap_x >= 1:
            return True

    return False

def check_circulation_graph(layout):
    """Validates connectivity using a Breadth-First Search graph traversal."""
    # Build graph adjacencies based on touching physical boundaries
    graph = {room: [] for room in ROOMS}
    for i in range(len(ROOMS)):
        for j in range(i + 1, len(ROOMS)):
            r1, r2 = ROOMS[i], ROOMS[j]
            if share_accessible_wall(layout[r1], layout[r2]):
                graph[r1].append(r2)
                graph[r2].append(r1)

    # BFS traversal starting from the central circulation spine ('Corridor')
    visited = set()
    queue = ['Corridor']

    if 'Corridor' in graph:
        visited.add('Corridor')
        while queue:
            current = queue.pop(0)
            for neighbor in graph[current]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)

    # Return count of isolated rooms that lack doors to the circulation graph
    unreachable_rooms = len(ROOMS) - len(visited)
    return unreachable_rooms

def check_daylight_compliance(room_rect, orientation):
    x, y, w, h = room_rect
    if orientation == 'North' and (y + h) >= SITE_HEIGHT: return True
    if orientation == 'South' and y <= 0: return True
    if orientation == 'East' and (x + w) >= SITE_WIDTH: return True
    if orientation == 'West' and x <= 0: return True
    return False

def calculate_fitness(layout):
    """Comprehensive multi-objective optimization function."""
    adjacency_score = 0
    overlap_penalty = 0
    daylight_reward = 0
    out_of_bounds_penalty = 0
    circulation_penalty = 0

    # 1. Adjacency Proxemics Tracking
    for i in range(len(ROOMS)):
        for j in range(i + 1, len(ROOMS)):
            r1, r2 = ROOMS[i], ROOMS[j]
            c1_x, c1_y = layout[r1][0] + layout[r1][2]/2, layout[r1][1] + layout[r1][3]/2
            c2_x, c2_y = layout[r2][0] + layout[r2][2]/2, layout[r2][1] + layout[r2][3]/2

            weight = get_adjacency_weight(r1, r2)
            if weight > 0:
                dist = math.sqrt((c1_x - c2_x)**2 + (c1_y - c2_y)**2)
                adjacency_score += weight / (dist + 0.1)

    # 2. Structural Wall Intersections & Boundary Envelopes
    for i, r1 in enumerate(ROOMS):
        x1, y1, w1, h1 = layout[r1]
        if x1 + w1 > SITE_WIDTH or y1 + h1 > SITE_HEIGHT or x1 < 0 or y1 < 0:
            out_of_bounds_penalty += 800

        for j, r2 in enumerate(ROOMS):
            if i != j:
                overlap = calculate_overlap_with_walls(layout[r1], layout[r2])
                if overlap > 0:
                    overlap_penalty += overlap * 300

    # 3. Solar Daylight Gain Orientation
    for room, orientation in DAYLIGHT_CONSTRAINTS.items():
        if check_daylight_compliance(layout[room], orientation):
            daylight_reward += 60

    # 4. Circulation Graph Validation Penalty
    isolated_count = check_circulation_graph(layout)
    if isolated_count > 0:
        circulation_penalty = isolated_count * 1000 # Critical safety rejection penalty

    total_fitness = adjacency_score + daylight_reward - overlap_penalty - out_of_bounds_penalty - circulation_penalty
    return max(0.01, total_fitness)

# --- 3. REPRODUCIBLE EVOLUTIONARY KERNEL ---
def generate_random_layout():
    layout = {}
    for room, bounds in ROOM_CONFIGS.items():
        min_w, max_w, min_h, max_h = bounds
        w = random.randint(min_w, max_w)
        h = random.randint(min_h, max_h)
        x = random.randint(0, max(0, SITE_WIDTH - w))
        y = random.randint(0, max(0, SITE_HEIGHT - h))
        layout[room] = [x, y, w, h]
    return layout

def select_parent(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    pick = random.uniform(0, total_fitness)
    current = 0
    for individual, score in zip(population, fitness_scores):
        current += score
        if current > pick:
            return individual
    return population[-1]

def crossover(parent1, parent2):
    child = {}
    split = random.randint(1, len(ROOMS) - 1)
    for i, room in enumerate(ROOMS):
        child[room] = list(parent1[room]) if i < split else list(parent2[room])
    return child

def mutate(layout):
    if random.random() < MUTATION_RATE:
        room = random.choice(ROOMS)
        mutation_type = random.choice(['pos', 'dim'])
        min_w, max_w, min_h, max_h = ROOM_CONFIGS[room]

        if mutation_type == 'pos':
            layout[room][0] = random.randint(0, max(0, SITE_WIDTH - layout[room][2]))
            layout[room][1] = random.randint(0, max(0, SITE_HEIGHT - layout[room][3]))
        else:
            layout[room][2] = random.randint(min_w, max_w)
            layout[room][3] = random.randint(min_h, max_h)
    return layout

def print_layout_matrix(layout):
    grid = [[" . " for _ in range(SITE_WIDTH)] for _ in range(SITE_HEIGHT)]

    # Paint spaces inside matrix
    for room, (x, y, w, h) in layout.items():
        char = room[0] if room != 'Bathroom' else 'b' # Map letters clean
        for r_y in range(y, min(y + h, SITE_HEIGHT)):
            for r_x in range(x, min(x + w, SITE_WIDTH)):
                grid[r_y][r_x] = f" {char} "

    for row in reversed(grid):
        print("".join(row))
    print(f"\nLegend: C=Corridor, L=Living Room, K=Kitchen, B=Bedroom, b=Bathroom")

# --- 4. EXECUTION PIPELINE ---
def run_structural_ga():
    population = [generate_random_layout() for _ in range(POPULATION_SIZE)]
    best_layout = None
    best_fitness = -1

    print(f"Evolving plan layout with structural wall bounds and connectivity graph logic...")

    for generation in range(GENERATIONS):
        fitness_scores = [calculate_fitness(ind) for ind in population]

        for ind, score in zip(population, fitness_scores):
            if score > best_fitness:
                best_fitness = score
                best_layout = ind

        next_population = []
        for _ in range(POPULATION_SIZE // 2):
            p1 = select_parent(population, fitness_scores)
            p2 = select_parent(population, fitness_scores)
            next_population.extend([mutate(crossover(p1, p2)), mutate(crossover(p2, p1))])
        population = next_population

    print(f"\nOptimization Finished. Score: {best_fitness:.2f}")
    isolated_rooms_left = check_circulation_graph(best_layout)
    print(f"Circulation Integrity Check: {isolated_rooms_left} isolated rooms detected.")

    print("\n--- STRUCTURAL MAP ---")
    print_layout_matrix(best_layout)


run_structural_ga()
```

```python
import random
import math
import json

# --- 1. CONFIGURATION & ARCHITECTURAL DATA ---
SITE_WIDTH = 12
SITE_HEIGHT = 12
FLOORS = [0, 1]  # 0 = Ground Floor, 1 = First Floor

POPULATION_SIZE = 40
GENERATIONS = 150
MUTATION_RATE = 0.35
WALL_THICKNESS = 1

# Rooms mapped to target floor placement constraints
ROOM_CONFIGS = {
    # 'RoomName': (min_w, max_w, min_h, max_h, allowed_floors)
    'Stairs':       (2, 2, 2, 2, [0, 1]),  # Must line up vertically
    'Living Room':  (4, 5, 4, 5, [0]),     # Ground only
    'Kitchen':      (3, 4, 3, 4, [0]),     # Ground only
    'Bed 1':        (3, 4, 3, 4, [1]),     # First floor only
    'Bed 2':        (3, 4, 3, 4, [1]),     # First floor only
    'Bathroom':     (2, 3, 2, 3, [0, 1])   # Allowed on any floor
}
ROOMS = list(ROOM_CONFIGS.keys())

ADJACENCY_MATRIX = {
    ('Living Room', 'Kitchen'): 5,
    ('Living Room', 'Stairs'): 4,
    ('Bed 1', 'Stairs'): 5,
    ('Bed 2', 'Stairs'): 5,
    ('Bed 1', 'Bed 2'): 3,
}

DAYLIGHT_CONSTRAINTS = {
    'Living Room': 'South',
    'Bed 1': 'East',
    'Kitchen': 'North'
}

def get_adjacency_weight(r1, r2):
    return ADJACENCY_MATRIX.get((r1, r2), ADJACENCY_MATRIX.get((r2, r1), 0))

# --- 2. ADVANCED 3D SPACE EVALUATION ENGINE ---
def calculate_overlap_3d(rect1, rect2):
    """Calculates overlaps including structural wall buffer if on the same floor level."""
    x1, y1, z1, w1, h1 = rect1
    x2, y2, z2, w2, h2 = rect2

    if z1 != z2:
        return 0  # No overlap penalty if rooms reside on separate floor plates

    x_left = max(x1 - WALL_THICKNESS, x2 - WALL_THICKNESS)
    y_top = max(y1 - WALL_THICKNESS, y2 - WALL_THICKNESS)
    x_right = min(x1 + w1 + WALL_THICKNESS, x2 + w2 + WALL_THICKNESS)
    y_bottom = min(y1 + h1 + WALL_THICKNESS, y2 + h2 + WALL_THICKNESS)

    if x_right > x_left and y_bottom > y_top:
        return (x_right - x_left) * (y_bottom - y_top)
    return 0

def check_vertical_alignment(layout):
    """Calculates alignment accuracy for vertical circulation shafts (Stairs)."""
    # Splitting stairs into floor instances for tracking
    stairs_g = layout.get('Stairs_F0')
    stairs_f1 = layout.get('Stairs_F1')

    if not stairs_g or not stairs_f1:
        return 0

    # Penalty based on offset distance between ground floor and first floor stairs
    dist_x = abs(stairs_g[0] - stairs_f1[0])
    dist_y = abs(stairs_g[1] - stairs_f1[1])
    return (dist_x + dist_y) * 1500  # Extreme alignment priority penalty

def check_daylight_compliance(room_rect, orientation):
    x, y, z, w, h = room_rect
    if orientation == 'North' and (y + h) >= SITE_HEIGHT: return True
    if orientation == 'South' and y <= 0: return True
    if orientation == 'East' and (x + w) >= SITE_WIDTH: return True
    if orientation == 'West' and x <= 0: return True
    return False

def calculate_fitness(layout):
    """Evaluates multi-story spatial layouts across voxel grids."""
    adjacency_score = 0
    overlap_penalty = 0
    daylight_reward = 0
    out_of_bounds_penalty = 0

    active_rooms = list(layout.keys())

    # 1. 3D Adjacency Mapping (Accounts for vertical floor offsets)
    for i in range(len(active_rooms)):
        for j in range(i + 1, len(active_rooms)):
            r1, r2 = active_rooms[i], active_rooms[j]
            orig_r1 = r1.split('_')[0]
            orig_r2 = r2.split('_')[0]

            weight = get_adjacency_weight(orig_r1, orig_r2)
            if weight > 0:
                # Target center points
                c1_x, c1_y, c1_z = layout[r1][0] + layout[r1][3]/2, layout[r1][1] + layout[r1][4]/2, layout[r1][2] * 3.0
                c2_x, c2_y, c2_z = layout[r2][0] + layout[r2][3]/2, layout[r2][1] + layout[r2][4]/2, layout[r2][2] * 3.0

                dist = math.sqrt((c1_x - c2_x)**2 + (c1_y - c2_y)**2 + (c1_z - c2_z)**2)
                adjacency_score += weight / (dist + 0.1)

    # 2. Structural Overlaps & Boundaries
    for i, r1 in enumerate(active_rooms):
        x1, y1, z1, w1, h1 = layout[r1]
        if x1 + w1 > SITE_WIDTH or y1 + h1 > SITE_HEIGHT or x1 < 0 or y1 < 0:
            out_of_bounds_penalty += 800

        for j, r2 in enumerate(active_rooms):
            if i != j:
                overlap = calculate_overlap_3d(layout[r1], layout[r2])
                if overlap > 0:
                    overlap_penalty += overlap * 400

    # 3. Daylight Optimization
    for room_inst, rect in layout.items():
        orig_name = room_inst.split('_')[0]
        if orig_name in DAYLIGHT_CONSTRAINTS:
            if check_daylight_compliance(rect, DAYLIGHT_CONSTRAINTS[orig_name]):
                daylight_reward += 60

    # 4. Vertical Structural Shaft Alignment Penalty
    vertical_penalty = check_vertical_alignment(layout)

    total_fitness = adjacency_score + daylight_reward - overlap_penalty - out_of_bounds_penalty - vertical_penalty
    return max(0.01, total_fitness)

# --- 3. REPRODUCIBLE EVOLUTIONARY FUNCTIONS ---
def generate_random_layout():
    """Generates an explicit multi-story room footprint mapping."""
    layout = {}
    for room, configs in ROOM_CONFIGS.items():
        min_w, max_w, min_h, max_h, allowed_floors = configs

        # Instantiate separate entries if room spans or is allowed across multiple floors
        for target_floor in allowed_floors:
            # unique identifier suffix instance for floor maps
            instance_name = f"{room}_F{target_floor}"
            w = random.randint(min_w, max_w)
            h = random.randint(min_h, max_h)
            x = random.randint(0, max(0, SITE_WIDTH - w))
            y = random.randint(0, max(0, SITE_HEIGHT - h))
            layout[instance_name] = [x, y, target_floor, w, h]
    return layout

def select_parent(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    pick = random.uniform(0, total_fitness)
    current = 0
    for individual, score in zip(population, fitness_scores):
        current += score
        if current > pick:
            return individual
    return population[-1]

def crossover(parent1, parent2):
    child = {}
    keys = list(parent1.keys())
    split = random.randint(1, len(keys) - 1)
    for i, key in enumerate(keys):
        child[key] = list(parent1[key]) if i < split else list(parent2[key])
    return child

def mutate(layout):
    if random.random() < MUTATION_RATE:
        room_key = random.choice(list(layout.keys()))
        orig_name = room_key.split('_')[0]
        min_w, max_w, min_h, max_h, _ = ROOM_CONFIGS[orig_name]

        if random.choice(['pos', 'dim']) == 'pos':
            layout[room_key][0] = random.randint(0, max(0, SITE_WIDTH - layout[room_key][3]))
            layout[room_key][1] = random.randint(0, max(0, SITE_HEIGHT - layout[room_key][4]))
        else:
            layout[room_key][3] = random.randint(min_w, max_w)
            layout[room_key][4] = random.randint(min_h, max_h)
    return layout

# --- 4. EXPORT PIPELINE FOR CAD TOOLS ---
def export_layout_to_json(layout, filename="architectural_layout.json"):
    """Saves multi-story building blocks directly into a readable standard exchange JSON file."""
    output_data = {
        "metadata": {
            "site_width": SITE_WIDTH,
            "site_height": SITE_HEIGHT,
            "wall_thickness": WALL_THICKNESS,
            "floor_height_multiplier": 3.0
        },
        "geometries": []
    }

    for room_instance, rect in layout.items():
        room_name, floor_id = room_instance.split('_')
        x, y, z, w, h = rect

        room_data = {
            "name": room_name,
            "floor": int(floor_id.replace('F', '')),
            "origin": {"x": x, "y": y, "z": z * 3.0}, # Multiplied by standard ceiling height
            "dimensions": {"width": w, "length": h, "height": 2.8}
        }
        output_data["geometries"].append(room_data)

    with open(filename, 'w') as json_file:
        json.dump(output_data, json_file, indent=4)
    print(f"\n[SUCCESS] CAD-Ready JSON Data exported successfully to: '{filename}'")

def print_multi_story_map(layout):
    """Renders floor plates side-by-side inside console layout."""
    for f in FLOORS:
        print(f"\n--- LAYER MAP: FLOOR LEVEL {f} ---")
        grid = [[" . " for _ in range(SITE_WIDTH)] for _ in range(SITE_HEIGHT)]
        for room_inst, (x, y, z, w, h) in layout.items():
            if z == f:
                char = room_inst[0] # Grab first signature letter
                for r_y in range(y, min(y + h, SITE_HEIGHT)):
                    for r_x in range(x, min(x + w, SITE_WIDTH)):
                        grid[r_y][r_x] = f" {char} "
        for row in reversed(grid):
            print("".join(row))

# --- 5. MAIN EXECUTION PIPELINE ---
def run_multistory_ga():
    population = [generate_random_layout() for _ in range(POPULATION_SIZE)]
    best_layout = None
    best_fitness = -1

    print(f"Evolving multi-story layout configurations with vertical alignment parameters...")

    for generation in range(GENERATIONS):
        fitness_scores = [calculate_fitness(ind) for ind in population]

        for ind, score in zip(population, fitness_scores):
            if score > best_fitness:
                best_fitness = score
                best_layout = ind

        next_population = []
        for _ in range(POPULATION_SIZE // 2):
            p1 = select_parent(population, fitness_scores)
            p2 = select_parent(population, fitness_scores)
            next_population.extend([mutate(crossover(p1, p2)), mutate(crossover(p2, p1))])
        population = next_population

    print(f"\nOptimization Finished. High Score: {best_fitness:.2f}")
    print_multi_story_map(best_layout)

    # Run data serialization file export step
    export_layout_to_json(best_layout)

run_multistory_ga()
```

 Rooms are rendered as semi-transparent solid 3D bounding boxes. The script adds structural gray padding outlines around each room to represent your structural wall thickness constraints, and colors the central Stairs spine distinctly to verify its vertical alignment across floor plates.

```python
import random
import math
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. CONFIGURATION & ARCHITECTURAL DATA ---
SITE_WIDTH = 12
SITE_HEIGHT = 12
FLOORS = [0, 1]  # 0 = Ground Floor, 1 = First Floor

POPULATION_SIZE = 40
GENERATIONS = 120
MUTATION_RATE = 0.35
WALL_THICKNESS = 1.0  # Structural margin around usable floor area

ROOM_CONFIGS = {
    'Stairs':       (2, 2, 2, 2, [0, 1]),  # Must line up vertically
    'Living Room':  (4, 5, 4, 5, [0]),     # Ground only
    'Kitchen':      (3, 4, 3, 4, [0]),     # Ground only
    'Bed 1':        (3, 4, 3, 4, [1]),     # First floor only
    'Bed 2':        (3, 4, 3, 4, [1]),     # First floor only
    'Bathroom':     (2, 3, 2, 3, [0, 1])   # Allowed on any floor
}
ROOMS = list(ROOM_CONFIGS.keys())

# Color Palette mapping for 3D UI Render Visualization
ROOM_COLORS = {
    'Stairs': '#E74C3C',       # Red Alert for safety infrastructure
    'Living Room': '#3498DB',   # Blue
    'Kitchen': '#F1C40F',       # Yellow
    'Bed 1': '#9B59B6',         # Purple
    'Bed 2': '#8E44AD',         # Deep Purple
    'Bathroom': '#2ECC71'       # Green
}

ADJACENCY_MATRIX = {
    ('Living Room', 'Kitchen'): 5,
    ('Living Room', 'Stairs'): 4,
    ('Bed 1', 'Stairs'): 5,
    ('Bed 2', 'Stairs'): 5,
}

DAYLIGHT_CONSTRAINTS = {'Living Room': 'South', 'Bed 1': 'East', 'Kitchen': 'North'}

def get_adjacency_weight(r1, r2):
    return ADJACENCY_MATRIX.get((r1, r2), ADJACENCY_MATRIX.get((r2, r1), 0))

# --- 2. EVALUATION & OPTIMIZATION KERNEL ---
def calculate_overlap_3d(rect1, rect2):
    x1, y1, z1, w1, h1 = rect1
    x2, y2, z2, w2, h2 = rect2
    if z1 != z2: return 0
    x_left = max(x1 - WALL_THICKNESS, x2 - WALL_THICKNESS)
    y_top = max(y1 - WALL_THICKNESS, y2 - WALL_THICKNESS)
    x_right = min(x1 + w1 + WALL_THICKNESS, x2 + w2 + WALL_THICKNESS)
    y_bottom = min(y1 + h1 + WALL_THICKNESS, y2 + h2 + WALL_THICKNESS)
    if x_right > x_left and y_bottom > y_top:
        return (x_right - x_left) * (y_bottom - y_top)
    return 0

def check_vertical_alignment(layout):
    stairs_g = layout.get('Stairs_F0')
    stairs_f1 = layout.get('Stairs_F1')
    if not stairs_g or not stairs_f1: return 0
    return (abs(stairs_g[0] - stairs_f1[0]) + abs(stairs_g[1] - stairs_f1[1])) * 1500

def check_daylight_compliance(room_rect, orientation):
    x, y, z, w, h = room_rect
    if orientation == 'North' and (y + h) >= SITE_HEIGHT: return True
    if orientation == 'South' and y <= 0: return True
    if orientation == 'East' and (x + w) >= SITE_WIDTH: return True
    if orientation == 'West' and x <= 0: return True
    return False

def calculate_fitness(layout):
    adjacency_score, overlap_penalty, daylight_reward, out_of_bounds_penalty = 0, 0, 0, 0
    active_rooms = list(layout.keys())

    for i in range(len(active_rooms)):
        for j in range(i + 1, len(active_rooms)):
            r1, r2 = active_rooms[i], active_rooms[j]
            weight = get_adjacency_weight(r1.split('_')[0], r2.split('_')[0])
            if weight > 0:
                c1_x, c1_y, c1_z = layout[r1][0] + layout[r1][3]/2, layout[r1][1] + layout[r1][4]/2, layout[r1][2] * 3.0
                c2_x, c2_y, c2_z = layout[r2][0] + layout[r2][3]/2, layout[r2][1] + layout[r2][4]/2, layout[r2][2] * 3.0
                dist = math.sqrt((c1_x - c2_x)**2 + (c1_y - c2_y)**2 + (c1_z - c2_z)**2)
                adjacency_score += weight / (dist + 0.1)

    for i, r1 in enumerate(active_rooms):
        x1, y1, z1, w1, h1 = layout[r1]
        if x1 + w1 > SITE_WIDTH or y1 + h1 > SITE_HEIGHT or x1 < 0 or y1 < 0:
            out_of_bounds_penalty += 800
        for j, r2 in enumerate(active_rooms):
            if i != j:
                overlap = calculate_overlap_3d(layout[r1], layout[r2])
                if overlap > 0: overlap_penalty += overlap * 400

    for room_inst, rect in layout.items():
        orig_name = room_inst.split('_')[0]
        if orig_name in DAYLIGHT_CONSTRAINTS:
            if check_daylight_compliance(rect, DAYLIGHT_CONSTRAINTS[orig_name]):
                daylight_reward += 60

    total_fitness = adjacency_score + daylight_reward - overlap_penalty - out_of_bounds_penalty - check_vertical_alignment(layout)
    return max(0.01, total_fitness)

# --- 3. REPRODUCIBLE GA OPERATORS ---
def generate_random_layout():
    layout = {}
    for room, configs in ROOM_CONFIGS.items():
        min_w, max_w, min_h, max_h, allowed_floors = configs
        for target_floor in allowed_floors:
            instance_name = f"{room}_F{target_floor}"
            w = random.randint(min_w, max_w)
            h = random.randint(min_h, max_h)
            x = random.randint(0, max(0, SITE_WIDTH - w))
            y = random.randint(0, max(0, SITE_HEIGHT - h))
            layout[instance_name] = [x, y, target_floor, w, h]
    return layout

def select_parent(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    pick = random.uniform(0, total_fitness)
    current = 0
    for individual, score in zip(population, fitness_scores):
        current += score
        if current > pick: return individual
    return population[-1]

def crossover(parent1, parent2):
    child = {}
    keys = list(parent1.keys())
    split = random.randint(1, len(keys) - 1)
    for i, key in enumerate(keys):
        child[key] = list(parent1[key]) if i < split else list(parent2[key])
    return child

def mutate(layout):
    if random.random() < MUTATION_RATE:
        room_key = random.choice(list(layout.keys()))
        orig_name = room_key.split('_')[0]
        min_w, max_w, min_h, max_h, _ = ROOM_CONFIGS[orig_name]
        if random.choice(['pos', 'dim']) == 'pos':
            layout[room_key][0] = random.randint(0, max(0, SITE_WIDTH - layout[room_key][3]))
            layout[room_key][1] = random.randint(0, max(0, SITE_HEIGHT - layout[room_key][4]))
        else:
            layout[room_key][3] = random.randint(min_w, max_w)
            layout[room_key][4] = random.randint(min_h, max_h)
    return layout

# --- 4. 2D/3D MATPLOTLIB VISUALIZATION ENGINE ---
def render_layout_3d(layout):
    """Generates an interactive 3D spatial plot showing interior zones and structural volumes."""
    fig = plt.figure(figsize=(12, 9))
    ax = fig.add_subplot(111, projection='3d')

    FLOOR_HEIGHT = 3.0  # Scale height scalar for realistic 3D volume ratios
    ROOM_HEIGHT = 2.7

    for room_instance, (x, y, z, w, h) in layout.items():
        room_name = room_instance.split('_')[0]
        base_color = ROOM_COLORS.get(room_name, '#95A5A6')

        # Calculate explicit Z level translation bounds
        z_min = z * FLOOR_HEIGHT
        z_max = z_min + ROOM_HEIGHT

        # Draw the main semi-transparent interior usable volume cube
        ax.bar3d(x, y, z_min, w, h, ROOM_HEIGHT,
                 color=base_color, alpha=0.55, edgecolor=base_color, linewidth=0.5)

        # Optional: Render Structural Wall Boundaries Shell Wireframe
        if WALL_THICKNESS > 0:
            # Convert hex to RGB and add alpha
            edge_rgb = tuple(int('#2C3E50'[i:i+2], 16) / 255 for i in (1, 3, 5))
            edge_rgba = edge_rgb + (0.15,) # Add alpha

            ax.bar3d(x - WALL_THICKNESS/2, y - WALL_THICKNESS/2, z_min,
                     w + WALL_THICKNESS, h + WALL_THICKNESS, ROOM_HEIGHT,
                     color=(0,0,0,0), edgecolor=edge_rgba, linewidth=1.0, linestyle=':')

        # Place centered label text within the 3D space blocks
        ax.text(x + w/2, y + h/2, z_min + ROOM_HEIGHT/2, room_name,
                color='#111111', fontsize=9, weight='bold', ha='center', va='center')

    # Set plot envelope configurations
    ax.set_xlim(0, SITE_WIDTH)
    ax.set_ylim(0, SITE_HEIGHT)
    ax.set_zlim(0, len(FLOORS) * FLOOR_HEIGHT)

    ax.set_xlabel('Site Width (X-Axis)')
    ax.set_ylabel('Site Length (Y-Axis)')
    ax.set_zlabel('Elevation Height (Z-Axis)')
    ax.set_title('Optimized Multi-Story Floor Plan (Genetic Algorithm 3D Model)', fontsize=14, weight='bold', pad=20)

    # Customize axis intervals to match floor boundaries explicitly
    ax.set_zticks([f * FLOOR_HEIGHT for f in FLOORS])
    ax.set_zticklabels([f"Floor {f}" for f in FLOORS])

    plt.tight_layout()
    plt.show()

# --- 5. MAIN PIPELINE ---
def main():
    population = [generate_random_layout() for _ in range(POPULATION_SIZE)]
    best_layout, best_fitness = None, -1

    print("Running Multi-Story Architecture GA Evolutionary Iteration Processing Loops...")
    for generation in range(GENERATIONS):
        fitness_scores = [calculate_fitness(ind) for ind in population]
        for ind, score in zip(population, fitness_scores):
            if score > best_fitness:
                best_fitness, best_layout = score, ind

        next_population = []
        for _ in range(POPULATION_SIZE // 2):
            p1, p2 = select_parent(population, fitness_scores), select_parent(population, fitness_scores)
            next_population.extend([mutate(crossover(p1, p2)), mutate(crossover(p2, p1))])
        population = next_population

    print(f"\nGenerative Space Generation Completed. Best Layout Fitness Value Match: {best_fitness:.2f}")
    print("Launching Matplotlib interactive spatial renderer canvas viewport...")

    # Fire up the 3D GUI canvas plot window
    render_layout_3d(best_layout)


main()
```

### Let us add a steady-state therml energy efficiency calculator to  evaluate structural sustainability of our layout.

- It evaluates the external envelope area of your 3D layout. If a room shares an edge with the outer site boundary (North, South, East, West) or constitutes the roof structure (top floor), the engine calculates its exposed surface square footage.

- It then applies standard insulation heat loss coefficients (\(U\)-values) to calculate total transmission heat loss in Watts per Kelvin (\(W/K\)).

Core Architectural Extensions

- Thermal Shell Tracker: Isolates walls and roofs exposed directly to exterior ambient air conditions versus internal partition walls.

- Material \(U\)-Value Mapping: Differentiates insulation performance benchmarks for external walls (\(U = 0.24\)) and roof structures (\(U = 0.16\)).

- Negative Fitness Feedback Loop: The total heat loss calculation (\(W/K\)) acts as a penalty metric. This encourages the algorithm to cluster spaces into compact, thermally efficient footprints (low surface-area-to-volume ratio) while balancing internal daylighting mandates.


```python
import random
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. CONFIGURATION & ARCHITECTURAL DATA ---
SITE_WIDTH = 12
SITE_HEIGHT = 12
FLOORS = [0, 1]  # 0 = Ground Floor, 1 = First Floor
FLOOR_HEIGHT = 3.0
ROOM_HEIGHT = 2.7

POPULATION_SIZE = 40
GENERATIONS = 150
MUTATION_RATE = 0.35
WALL_THICKNESS = 1.0

# Thermal Physics Properties (U-Values: W/m²K - Lower is better insulated)
U_EXT_WALL = 0.24  # Standard high-efficiency external insulation wall
U_ROOF = 0.16      # High performance roof insulation barrier

ROOM_CONFIGS = {
    'Stairs':       (2, 2, 2, 2, [0, 1]),
    'Living Room':  (4, 5, 4, 5, [0]),
    'Kitchen':      (3, 4, 3, 4, [0]),
    'Bed 1':        (3, 4, 3, 4, [1]),
    'Bed 2':        (3, 4, 3, 4, [1]),
    'Bathroom':     (2, 3, 2, 3, [0, 1])
}
ROOMS = list(ROOM_CONFIGS.keys())

ROOM_COLORS = {
    'Stairs': '#E74C3C', 'Living Room': '#3498DB', 'Kitchen': '#F1C40F',
    'Bed 1': '#9B59B6', 'Bed 2': '#8E44AD', 'Bathroom': '#2ECC71'
}

ADJACENCY_MATRIX = {
    ('Living Room', 'Kitchen'): 5,
    ('Living Room', 'Stairs'): 4,
    ('Bed 1', 'Stairs'): 5,
    ('Bed 2', 'Stairs'): 5,
}

DAYLIGHT_CONSTRAINTS = {'Living Room': 'South', 'Bed 1': 'East'}

def get_adjacency_weight(r1, r2):
    return ADJACENCY_MATRIX.get((r1, r2), ADJACENCY_MATRIX.get((r2, r1), 0))

# --- 2. ENERGY EFFICIENCY & CALCULATION ENGINE ---
def calculate_thermal_heat_loss(layout):
    """Calculates transmission heat loss (W/K) across exposed external building envelopes."""
    total_heat_loss = 0.0

    for room_instance, (x, y, z, w, h) in layout.items():
        exposed_wall_area = 0.0
        exposed_roof_area = 0.0

        # Check exposure along the 4 primary cardinal orientations of the site envelope
        if y <= 0:                          # South Exposure
            exposed_wall_area += (w * ROOM_HEIGHT)
        if (y + h) >= SITE_HEIGHT:          # North Exposure
            exposed_wall_area += (w * ROOM_HEIGHT)
        if x <= 0:                          # West Exposure
            exposed_wall_area += (h * ROOM_HEIGHT)
        if (x + w) >= SITE_WIDTH:           # East Exposure
            exposed_wall_area += (h * ROOM_HEIGHT)

        # Check roof envelope exposure (if room is located on top structural tier plate)
        if z == max(FLOORS):
            exposed_roof_area += (w * h)

        # Fourier's Law of Heat Conduction: Q = U * A * dT (Calculated per Kelvin degree gradient)
        room_loss = (exposed_wall_area * U_EXT_WALL) + (exposed_roof_area * U_ROOF)
        total_heat_loss += room_loss

    return total_heat_loss

# --- 3. GEOMETRIC INTERSECTION LAYER ---
def calculate_overlap_3d(rect1, rect2):
    x1, y1, z1, w1, h1 = rect1
    x2, y2, z2, w2, h2 = rect2
    if z1 != z2: return 0
    x_left = max(x1 - WALL_THICKNESS, x2 - WALL_THICKNESS)
    y_top = max(y1 - WALL_THICKNESS, y2 - WALL_THICKNESS)
    x_right = min(x1 + w1 + WALL_THICKNESS, x2 + w2 + WALL_THICKNESS)
    y_bottom = min(y1 + h1 + WALL_THICKNESS, y2 + h2 + WALL_THICKNESS)
    if x_right > x_left and y_bottom > y_top:
        return (x_right - x_left) * (y_bottom - y_top)
    return 0

def check_vertical_alignment(layout):
    stairs_g = layout.get('Stairs_F0')
    stairs_f1 = layout.get('Stairs_F1')
    if not stairs_g or not stairs_f1: return 0
    return (abs(stairs_g[0] - stairs_f1[0]) + abs(stairs_g[1] - stairs_f1[1])) * 1500

def check_daylight_compliance(room_rect, orientation):
    x, y, z, w, h = room_rect
    if orientation == 'North' and (y + h) >= SITE_HEIGHT: return True
    if orientation == 'South' and y <= 0: return True
    if orientation == 'East' and (x + w) >= SITE_WIDTH: return True
    if orientation == 'West' and x <= 0: return True
    return False

def calculate_fitness(layout):
    adjacency_score, overlap_penalty, daylight_reward, out_of_bounds_penalty = 0, 0, 0, 0
    active_rooms = list(layout.keys())

    for i in range(len(active_rooms)):
        for j in range(i + 1, len(active_rooms)):
            r1, r2 = active_rooms[i], active_rooms[j]
            weight = get_adjacency_weight(r1.split('_')[0], r2.split('_')[0])
            if weight > 0:
                c1_x, c1_y, c1_z = layout[r1][0] + layout[r1][3]/2, layout[r1][1] + layout[r1][4]/2, layout[r1][2] * FLOOR_HEIGHT
                c2_x, c2_y, c2_z = layout[r2][0] + layout[r2][3]/2, layout[r2][1] + layout[r2][4]/2, layout[r2][2] * FLOOR_HEIGHT
                dist = math.sqrt((c1_x - c2_x)**2 + (c1_y - c2_y)**2 + (c1_z - c2_z)**2)
                adjacency_score += weight / (dist + 0.1)

    for i, r1 in enumerate(active_rooms):
        x1, y1, z1, w1, h1 = layout[r1]
        if x1 + w1 > SITE_WIDTH or y1 + h1 > SITE_HEIGHT or x1 < 0 or y1 < 0:
            out_of_bounds_penalty += 800
        for j, r2 in enumerate(active_rooms):
            if i != j:
                overlap = calculate_overlap_3d(layout[r1], layout[r2])
                if overlap > 0: overlap_penalty += overlap * 400

    for room_inst, rect in layout.items():
        orig_name = room_inst.split('_')[0]
        if orig_name in DAYLIGHT_CONSTRAINTS:
            if check_daylight_compliance(rect, DAYLIGHT_CONSTRAINTS[orig_name]):
                daylight_reward += 60

    # Integration of Thermal Envelope Efficiency Penalty
    heat_loss_w_k = calculate_thermal_heat_loss(layout)
    thermal_efficiency_penalty = heat_loss_w_k * 5.0  # Scale multiplier factor

    total_fitness = adjacency_score + daylight_reward - overlap_penalty - out_of_bounds_penalty - check_vertical_alignment(layout) - thermal_efficiency_penalty
    return max(0.01, total_fitness)

# --- 4. BREEDING GENERATORS ---
def generate_random_layout():
    layout = {}
    for room, configs in ROOM_CONFIGS.items():
        min_w, max_w, min_h, max_h, allowed_floors = configs
        for target_floor in allowed_floors:
            instance_name = f"{room}_F{target_floor}"
            w = random.randint(min_w, max_w)
            h = random.randint(min_h, max_h)
            x = random.randint(0, max(0, SITE_WIDTH - w))
            y = random.randint(0, max(0, SITE_HEIGHT - h))
            layout[instance_name] = [x, y, target_floor, w, h]
    return layout

def select_parent(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    pick = random.uniform(0, total_fitness)
    current = 0
    for individual, score in zip(population, fitness_scores):
        current += score
        if current > pick: return individual
    return population[-1]

def crossover(parent1, parent2):
    child = {}
    keys = list(parent1.keys())
    split = random.randint(1, len(keys) - 1)
    for i, key in enumerate(keys):
        child[key] = list(parent1[key]) if i < split else list(parent2[key])
    return child

def mutate(layout):
    if random.random() < MUTATION_RATE:
        room_key = random.choice(list(layout.keys()))
        orig_name = room_key.split('_')[0]
        min_w, max_w, min_h, max_h, _ = ROOM_CONFIGS[orig_name]
        if random.choice(['pos', 'dim']) == 'pos':
            layout[room_key][0] = random.randint(0, max(0, SITE_WIDTH - layout[room_key][3]))
            layout[room_key][1] = random.randint(0, max(0, SITE_HEIGHT - layout[room_key][4]))
        else:
            layout[room_key][3] = random.randint(min_w, max_w)
            layout[room_key][4] = random.randint(min_h, max_h)
    return layout

# --- 5. RENDER SYSTEM ---
def render_layout_3d(layout, thermal_rating):
    fig = plt.figure(figsize=(12, 9))
    ax = fig.add_subplot(111, projection='3d')

    for room_instance, (x, y, z, w, h) in layout.items():
        room_name = room_instance.split('_')[0]
        base_color = ROOM_COLORS.get(room_name, '#95A5A6')
        z_min = z * FLOOR_HEIGHT

        ax.bar3d(x, y, z_min, w, h, ROOM_HEIGHT, color=base_color, alpha=0.55, edgecolor=base_color, linewidth=0.5)
        ax.text(x + w/2, y + h/2, z_min + ROOM_HEIGHT/2, room_name, color='#111111', fontsize=9, weight='bold', ha='center', va='center')

    ax.set_xlim(0, SITE_WIDTH)
    ax.set_ylim(0, SITE_HEIGHT)
    ax.set_zlim(0, len(FLOORS) * FLOOR_HEIGHT)
    ax.set_zticks([f * FLOOR_HEIGHT for f in FLOORS])
    ax.set_zticklabels([f"Floor {f}" for f in FLOORS])

    title_str = f'Optimized Layout Plan\nEnvelope Transmission Heat Loss: {thermal_rating:.2f} W/K'
    ax.set_title(title_str, fontsize=12, weight='bold', pad=10)
    plt.show()

# --- 6. EXECUTION PIPELINE ---
def main():
    population = [generate_random_layout() for _ in range(POPULATION_SIZE)]
    best_layout, best_fitness = None, -1

    for generation in range(GENERATIONS):
        fitness_scores = [calculate_fitness(ind) for ind in population]
        for ind, score in zip(population, fitness_scores):
            if score > best_fitness:
                best_fitness, best_layout = score, ind

        next_population = []
        for _ in range(POPULATION_SIZE // 2):
            p1, p2 = select_parent(population, fitness_scores), select_parent(population, fitness_scores)
            next_population.extend([mutate(crossover(p1, p2)), mutate(crossover(p2, p1))])
        population = next_population

    final_heat_loss = calculate_thermal_heat_loss(best_layout)
    print(f"\nGenerative Run Complete. Optimal Building Heat Loss Calculated: {final_heat_loss:.2f} W/K")
    render_layout_3d(best_layout, final_heat_loss)


main()
```

#### Architectural Impact of the Thermal Efficiency Calculator

- Clustering Behavior: Because open borders drop fitness points through thermal_efficiency_penalty, the algorithm naturally learns to stack and nest modules tightly, minimizing long, exposed corridors or sprawling footprints.

- Passive Zoning: Daylighting constraints (e.g., placing the Living Room on the South facade) conflict with heat loss constraints. The GA works to locate rooms on exterior walls where required for windows, while drawing non-daylight zones (like Bathrooms and Stairs) deep into the structural core to form insulation buffers.
