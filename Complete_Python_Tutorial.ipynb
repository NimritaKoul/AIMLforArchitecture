### 2.1 The 'While' loop
A while loop in Python is a control flow statement that repeatedly executes a block of code as long as a specified condition evaluates to True. It is used when the number of iterations is not known beforehand and depends on the condition being met.

```
while condition:
    # Code block to execute
```

- condition: A boolean expression that is evaluated before each iteration. If True, the code block runs; if False, the loop terminates.

- The code block (indented under the while statement) executes repeatedly until the condition becomes False.

Key Features

- The loop continues indefinitely if the condition never becomes False, potentially causing an infinite loop unless managed (e.g., with a break statement or condition update).

- The condition is checked before each iteration, so the loop may not execute at all if the condition is initially False.

- Python supports loop control statements:
    - break: Exits the loop immediately.
    - continue: Skips the rest of the current iteration and checks the condition again.
    - else: Executes a block of code when the condition becomes False, unless the loop was exited with break.
- Use correct indentation to (usually 4 spaces) to define the code block within the loop. 
- Infinite Loop Risk: Ensure the condition will eventually become False or use break to avoid infinite loops.
- While loop is ideal for situations where the loop’s termination depends on dynamic conditions (e.g., user input, external data, or calculations)

````python
a = 0
while a < 10:
    a = a + 1
    print(a)
````

## 2.2 Boolean Expressions (True or False)

A Boolean expression in Python is an expression that evaluates to either True or False, the two possible values of Python’s bool data type. These expressions are fundamental in controlling program flow, especially in constructs like if statements, while loops, and logical operations. They are typically formed using comparison operators, logical operators, or other operations that yield a Boolean result.

### Key Components of Boolean Expressions

1. **Comparison Operators**:
   These compare two values and return `True` or `False` based on the relationship.
   - `==`: Equal to (e.g., `5 == 5` → `True`)
   - `!=`: Not equal to (e.g., `5 != 3` → `True`)
   - `<`: Less than (e.g., `3 < 5` → `True`)
   - `>`: Greater than (e.g., `5 > 3` → `True`)
   - `<=`: Less than or equal to (e.g., `4 <= 4` → `True`)
   - `>=`: Greater than or equal to (e.g., `6 >= 5` → `True`)

2. **Logical Operators**:
   These combine or modify Boolean expressions.
   - `and`: Returns `True` if both operands are `True` (e.g., `(5 > 3) and (2 < 4)` → `True`)
   - `or`: Returns `True` if at least one operand is `True` (e.g., `(5 < 3) or (2 < 4)` → `True`)
   - `not`: Negates the Boolean value (e.g., `not (5 > 3)` → `False`)

3. **Membership Operators**:
   Test for membership in a sequence (e.g., list, string, tuple).
   - `in`: Returns `True` if a value is in the sequence (e.g., `3 in [1, 2, 3]` → `True`)
   - `not in`: Returns `True` if a value is not in the sequence (e.g., `4 not in [1, 2, 3]` → `True`)

4. **Identity Operators**:
   Check if two variables refer to the same object in memory.
   - `is`: Returns `True` if both operands are the same object (e.g., `a is b` → `True` if `a` and `b` point to the same object)
   - `is not`: Returns `True` if operands are not the same object (e.g., `a is not b`)

5. **Boolean Values and Truthiness**:
   - Direct Boolean values: `True` and `False`.
   - Python evaluates certain values as `True` or `False` in a Boolean context (truthiness):
     - **Falsy** values: `False`, `None`, `0`, `0.0`, empty strings (`""`), empty lists (`[]`), empty tuples (`()`), empty dictionaries (`{}`), etc.
     - **Truthy** values: Non-zero numbers, non-empty strings, non-empty collections, etc.
     - Example: `if "hello":` evaluates to `True` because non-empty strings are truthy.

### **Examples of Boolean Expressions**

1. **Simple Comparison**:
   ```python
   x = 10
   y = 5
   print(x > y)  # True
   print(x == y)  # False
   ```

2. **Logical Operators**:
   ```python
   a = 3
   b = 7
   print((a < 5) and (b > 6))  # True (both conditions are True)
   print((a > 5) or (b > 6))   # True (one condition is True)
   print(not (a == 3))          # False (negates True)
   ```

3. **Membership Testing**:
   ```python
   fruits = ["apple", "banana", "cherry"]
   print("banana" in fruits)     # True
   print("orange" not in fruits) # True
   ```

4. **Identity Testing**:
   ```python
   a = [1, 2, 3]
   b = a
   c = [1, 2, 3]
   print(a is b)    # True (same object)
   print(a is c)    # False (different objects, even if contents are the same)
   ```

5. **Truthiness**:
   ```python
   name = "Python"
   empty_list = []
   print(bool(name))        # True (non-empty string is truthy)
   print(bool(empty_list))  # False (empty list is falsy)
   ```

6. **In a While Loop** (relating to your previous question):
   ```python
   count = 1
   while count <= 5:  # Boolean expression: count <= 5
       print(count)
       count += 1
   ```
   - The loop runs as long as `count <= 5` evaluates to `True`.

### **Important**
- **Evaluation**: Boolean expressions are evaluated from left to right, with short-circuiting for `and` and `or` (e.g., if the left side of `and` is `False`, the right side isn’t evaluated).
- **Use in Control Flow**: Boolean expressions are critical in `if`, `elif`, `while`, and `for` loops to control program execution.
- **Type Safety**: The result of a Boolean expression is always a `bool` (`True` or `False`), unless explicitly converted (e.g., `int(True)` → `1`).
- **Precedence**: Operators have precedence (e.g., `and` has lower precedence than `==`). Use parentheses to clarify complex expressions (e.g., `(a > b) and (c < d)`).
- Don't get `=` and `==` mixed up - the `=` operator makes what is on the left equal to what is on the right. the `==` operator says whether the thing on the left is the same as what is on the right, and returns True or False.


## 2.3 Conditional Statements

Conditional statements in Python are control flow structures that allow a program to execute specific blocks of code based on whether a given Boolean expression evaluates to True or False. They are used to make decisions in a program, enabling it to perform different actions depending on conditions. Python’s primary conditional statements are if, elif, and else.

Syntax:

```
if condition:
    # Code block executed if condition is True
elif another_condition:  # Optional, can have multiple elif clauses
    # Code block executed if another_condition is True
else:  # Optional, runs if all conditions are False
    # Code block executed if no conditions are True
```
Here: 
- condition: A Boolean expression (e.g., x > 5) that evaluates to True or False.
- Code block: Indented lines (typically 4 spaces) under each clause that execute if the corresponding condition is True.
- elif: Short for "else if," used to check additional conditions if the previous ones are False.
- else: Catches all cases where no previous conditions are True.

### 2.3.1 Key Features

- Indentation: Python uses indentation to define the scope of code blocks under if, elif, and else. Incorrect indentation raises an error.
- Boolean Expressions: Conditions are typically comparisons (e.g., ==, <, >) or logical operations (e.g., and, or, not) that return True or False.
- Truthiness: Python evaluates non-Boolean values in conditions (e.g., non-zero numbers, non-empty strings are True; 0, None, empty sequences are False).
- Single-Line Conditional: Python supports concise conditional expressions (ternary operators) for simple cases.

### 2.3.2 Types of Conditional Statements
- Simple if Statement: Executes a block of code if the condition is True.
- ** if-else Statement**:Provides an alternative block if the condition is False.
- ** if-elif-else Statement**: Checks multiple conditions sequentially, executing the first True condition’s block or the else block if none are True.
- Nested Conditionals:Conditions within conditions for more complex logic
- Conditional Expression (Ternary Operator):A concise way to write simple if-else logic in one line.

`else` simply tells the computer what to do if the conditions of `if` aren't met. For example, read the following:

````python
a = 1
if a > 5:
    print("This shouldn't happen.")
else:
    print("This should happen.")
````

`a` is not greater than five, therefore what is under `else` is done.

`elif` is just a shortened way of saying `else if`. When the `if` statement fails to be true, `elif` will do what is under it IF the conditions are met. For example:

````python
z = 4
if z > 70:
    print("Something is very wrong")
elif z < 7:
    print("This is normal")
````

## 3. Indentation

# 4. Containers: Tuples, Lists, and Dictionaries

In Python, a container is a data structure that can hold and manage a collection of objects (values or variables). It "contains" multiple items and provides ways to access, iterate, and manipulate them.

I.e., while normal Python variables store one piece of information that may change over time, containers store multiple pieces of information.

- Containers can store homogeneous (same type) or heterogeneous (different type) data.

- They support operations like indexing, slicing, iteration, membership testing, and length checking.

- Many containers are part of Python’s built-in data types

## Basic containers types

> Note: **mutable** objects can be modified after creation and **immutable** objects cannot.

Containers are objects that can be used to group other objects together. The basic container types include:

- **`str`** (string: immutable; indexed by integers; items are stored in the order they were added)
- **`list`** (list: mutable; indexed by integers; items are stored in the order they were added)
  - `[3, 5, 6, 3, 'dog', 'cat', False]`

  - `(3, 5, 6, 3, 'dog', 'cat', False)`

  - `{3, 5, 6, 3, 'dog', 'cat', False}`

When defining lists, tuples, or sets, use commas (,) to separate the individual items. When defining dicts, use a colon (:) to separate keys from values and commas (,) to separate the key-value pairs.

## 4.4 Sets
A set in Python is an unordered collection of unique, immutable elements.
It is useful when you need to store items without duplicates and perform mathematical set operations (union, intersection, difference, etc.).

Key points about sets:

- Defined using curly braces { } or the set() constructor.

- Unordered → elements do not have a fixed position.

- No duplicates allowed → duplicates are automatically removed.

- Mutable → you can add or remove elements.

- Elements must be immutable (numbers, strings, tuples), but the set itself is mutable.

````python
# Creating sets
s1 = {1, 2, 3, 3, 4}
print(s1)   # {1, 2, 3, 4} (duplicate removed)

s2 = set([4, 5, 6])  # from list

# Adding and removing elements
s1.add(5)         # {1, 2, 3, 4, 5}
s1.remove(2)      # {1, 3, 4, 5}

# Set operations
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))        # {1, 2, 3, 4, 5}
print(a.intersection(b)) # {3}
print(a.difference(b))   # {1, 2}
````

| **Purpose**          | Acts as a placeholder for the value the function expects. | Supplies the real data to the function.                        |
| **Example**          | `def add(a, b):` → `a` and `b` are parameters             | `add(5, 3)` → `5` and `3` are arguments                        |

We want a menu that is returned to every time you finish adding, subtracting, etc. In other words, we will use a while loop.

````python
#calculator program
#this variable tells the loop whether it should loop or not.
#1 means loop. Anything else means don't loop.
loop = 1
#this variable holds the user's choice in the menu:
choice = 0
while loop == 1:
    #print what options you have
    print("Welcome to calculator.py")
    print("your options are:")
    print(" ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Quit calculator.py")
    print(" ")
    choice = eval(input("Choose your option: "))
    if choice == 1:
        add1 = eval(input("Add this: "))
        add2 = eval(input("to this: "))
        print(add1, "+", add2, "=", add1 + add2)
    elif choice == 2:
        sub2 = eval(input("Subtract this: "))
        sub1 = eval(input("from this: "))
        print(sub1, "-", sub2, "=", sub1 - sub2)
    elif choice == 3:
        mul1 = eval(input("Multiply this: "))
        mul2 = eval(input("with this: "))
        print(mul1, "*", mul2, "=", mul1 * mul2)
    elif choice == 4:
        div1 = eval(input("Divide this: "))
        div2 = eval(input("by this: "))
        print(div1, "/", div2, "=", div1 / div2)
    elif choice == 5:
        loop = 0   
print("Thank you for using calculator.py!")
````

### 5.6.2 Let us redesign our calculator program using functions:

````python
# calculator program
def menu():
    #print what options you have
    print("Welcome to calculator.py")
    print("your options are:")
    print(" ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Quit calculator.py")
    print(" ")
    return eval(input("Choose your option: "))
    
# this adds two numbers given
def add(a,b):
    print(a, "+", b, "=", a + b)
    
# this subtracts two numbers given
def sub(a,b):
    print(b, "-", a, "=", b - a)
    
# this multiplies two numbers given
def mul(a,b):
    print(a, "*", b, "=", a * b)
    
# this divides two numbers given
def div(a,b):
    print(a, "/", b, "=", a / b)
    
# NOW THE PROGRAM REALLY STARTS, AS CODE IS RUN
loop = 1
choice = 0
while loop == 1:
    choice = menu()
    if choice == 1:
        add(eval(input("Add this: ")),eval(input("to this: ")))
    elif choice == 2:
        sub(eval(input("Subtract this: ")),eval(input("from this: ")))
    elif choice == 3:
        mul(eval(input("Multiply this: ")),eval(input("by this: ")))
    elif choice == 4:
        div(eval(input("Divide this: ")),eval(input("by this: ")))
    elif choice == 5:
        loop = 0

print("Thank you for using calculator.py!")
````

##  6.1 Creating a `Class`
We can create a class in Python using the keyword 'class' and following syntax:

## 6.2 Python object attributes (methods and properties)

## 6.3 Class Attributes and Methods

### Class Attributes
- Class Attributes are the variables that belong to the class itself, rather than any particular object (instance) of the class.
- All instances of the class share the same value of class attributes.
- They are declared inside the class but outside any methods.

### Methods

Methods are the functions defined inside a class that define the behavior of the class objects.

Types of methods:

- Instance methods → operate on instance attributes; first parameter is self.
- Class methods → operate on class attributes; decorated with @classmethod and first parameter is cls.
- Static methods → don’t access instance or class attributes; decorated with @staticmethod.

```
class Student:
    school = "ABC School"  # class attribute

    def __init__(self, name):
        self.name = name   # instance attribute

    def greet(self):        # instance method
        print(f"Hello, I am {self.name} from {Student.school}")

    @classmethod
    def change_school(cls, new_school):  # class method
        cls.school = new_school

    @staticmethod
    def info():              # static method
        print("Students attend school.")

# Usage
s1 = Student("Ram")
s1.greet()           # Hello, I am Ram from ABC School
Student.change_school("XYZ School")
s1.greet()           # Hello, I am Ram from XYZ School
Student.info()       # Students attend school.

```

# 6.7 Object Oriented Programming in Python

Object-Oriented Programming (OOP) in Python is a programming paradigm that uses objects—instances of classes—to model real-world entities and organize code into reusable, modular structures. OOP focuses on encapsulating data (attributes) and behavior (methods) within objects, enabling concepts like inheritance, polymorphism, and encapsulation to create flexible and maintainable programs.

## 6.7.1Key Concepts of OOP in Python
OOP in Python revolves around four main principles, supported by its syntax for defining classes and objects:

1. Encapsulation: Bundling data (attributes) and methods (functions) that operate on that data into a single unit (class). Controls access to data using access modifiers (e.g., private attributes with _ or __ prefixes).

```Python
class Person:
    def __init__(self, name):
        self._name = name  # Protected attribute
    
    def get_name(self):
        return self._name

p = Person("Alice")
print(p.get_name())  # Output: Alice
```


2. Inheritance: Allows a class (child/subclass) to inherit attributes and methods from another class (parent/superclass), promoting code reuse. Supports single, multiple, and multilevel inheritance.Example:

```Python
class Animal:
    def speak(self):
        return "I make a sound"

class Dog(Animal):  # Inherits from Animal
    def speak(self):  # Override method
        return "Woof!"

dog = Dog()
print(dog.speak())  # Output: Woof!
```

3. Polymorphism: Allows different classes to be treated as instances of a common superclass, with methods behaving differently based on the object calling them. Achieved through method overriding or method overloading (via default arguments in Python). Example:
```Python
class Cat(Animal):
    def speak(self):
        return "Meow!"

animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())  # Output: Woof!, Meow!
```

4. Abstraction: Hides complex implementation details and exposes only necessary functionality. Achieved using abstract base classes (from the abc module). Example:
```Python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area())  # Output: 78.5

```

* A class is in the same category of things as variables, lists, dictionaries, etc. That is, they are *objects*
* A class is known as a 'data structure' - it holds data, and the methods to process that data.

A module in Python is a file containing Python code (functions, classes, variables, or runnable statements) that can be imported and used in other Python programs.

Modules help in organizing code, reusing functionality, and maintaining large programs.


Key points about modules:

- A module is simply a .py file.

- Can contain:

    - Functions

    - Classes

    -  Variables

    -  Executable code

- Can be imported using the import statement.

- Python has a large standard library of modules (math, os, sys, random, etc.).

- You can also create user-defined modules.



For example, a module might look like this, which we store in a file `moduletest.py`:
```Python
### EXAMPLE PYTHON MODULE
# Define some variables:
numberone = 1
ageofqueen = 78

# define some functions
def printhello():
    print("hello")

    print(eval(input) * 4)

        self.type = input("What type of piano? ")
        self.height = input("What height (in feet)? ")
        self.price = input("How much did it cost? ")
        self.age = input("How old is it (in years)? ")
	
    def printdetails(self):
        print("This piano is a/an " + self.height + " foot", end=" ")
        print(self.type, "piano, " + self.age, "years old and costing\
         " + self.price + " dollars.")
```

A module looks like your normal Python program.

## 7.1 importing modules
We `import` bits of it (or all of it) into other programs.

To import all the variables, functions and classes from `moduletest.py` into another program you are writing, we use the `import` operator. For example, to import `moduletest.py` into your main program (`mainprogram.py`), you would have this:

```Python
### mainprogam.py
### IMPORTS ANOTHER MODULE
import moduletest
```

This assumes that the module is in the same directory as `mainprogram.py`, or is a default module that comes with Python. You leave out the `.py` at the end of the file name - it is ignored. You normally put all `import` statements at the beginning of the Python file, but technically they can be anywhere. In order to use the items in the module in your main program, you use the following:

```Python
### USING AN IMPORTED MODULE
# Use the form modulename.itemname
# Examples:
print(moduletest.ageofqueen)
cfcpiano = moduletest.Piano()
cfcpiano.printdetails()
```

As you see, the modules that you import act very much like the classes we looked at last lesson - anything inside them must be preceded with `modulename.` for it to work.

##  7.2 Import specific classes or parts of a module

Wish you could get rid of the `modulename.` part that you have to put before every item you use from a module? One way to avoid this hassle is to import only the wanted objects from the module. To do this, you use the `from` operator. You use it in the form of `from modulename import itemname`. Here is an example:

```Python
### IMPORT ITEMS DIRECTLY INTO YOUR PROGRAM

# import them
from moduletest import ageofqueen
from moduletest import printhello

# now try using them
print(ageofqueen)
printhello()
```

This makes code more readable. 

If you wanted to, you could import everything from a module in this way by using `from modulename import *`. Of course, this can be troublesome if there are objects in your program with the same name as some items in the module. With large modules, this can easily happen, and can cause many a headache. A better way to do this would be to import a module in the normal way (without the `from` operator) and then assign items to a local name:

```Python
### ASSIGNING ITEMS TO A LOCAL NAME

# Assigning to a local name
timesfour = moduletest.timesfour

# Using the local name
print(timesfour(565))
```

This way, you can remove some crypticness, AND have all of the items from a certain module.

A final handy way to import modules is with an alias. Maybe you want to change a name because you've already used the same name for something else in your program, another module you imported uses the same name, or maybe you want to abbreviate a longer name that you use a lot. We can then use the `as` operator. That looks like this:

```Python
### IMPORT A MODULE WITH AN ALIAS
# import module
import moduletest as mt

# use module
print(mt.age)
cfcpiano = mt.Piano()
cfcpiano.printdetails()
```

With modules, it is incredibly easy to make programs that can grow in complexity without ending up with one cryptic file that is full of bugs. Modules are great for importing code.

## 8.1 Opening a file

You pass certain parameters to `open()` to tell it in which way the file should be opened - `r` for read only, `w` for writing only (if there is an old file, it will be written over), `a` for appending (adding things on to the end of the file) and `r+` for both reading and writing.

filename = 'C:\\temp\\readme.txt'

## 8.2 Seek 
In Python, the seek() method is used in file handling to move the file pointer (cursor) to a specific position within the file. The file pointer determines where the next read or write operation will start.

The cursor tells the read function (and many other I/O functions) where to start from. To set where the cursor is, you use the `seek()` function. It is used in the form `seek(offset, whence)`. 

`whence` is optional, and determines where to seek from. If `whence` is 0, the bytes/letters are counted from the beginning. If it is 1, the bytes are counted from the current cursor position. If it is 2, then the bytes are counted from the end of the file. If nothing is put there, 0 is assumed.

`offset` decribes how far from `whence` that the cursor moves. for example:
* `fl.seek(45,0)` would move the cursor to 45 bytes/letters after the beginning of the file.
* `fl.seek(10,1)` would move the cursor to 10 bytes/letters after the current cursor position.
* `fl.seek(-77,2)` would move the cursor to 77 bytes/letters before the end of the file (notice the - before the 77)

We can use `fl.seek()` to go to any spot in the file and then try typing `print(fl.read())`. It will print from the spot you seeked to. But realise that `fl.read()` moves the cursor to the end of the file - you will have to seek again.


```Python
# Create a sample file
with open("sample.txt", "w") as f:
    f.write("Hello, Python!")
# Open file for reading
with open("sample.txt", "r") as f:
    print(f.read(5))   # Output: Hello
    f.seek(7)          # Move pointer to 7th byte
    print(f.read())    # Output: Python!
```

- f.read(5) reads first 5 characters.
- f.seek(7) moves the pointer to the 8th character (index 7).
- Next f.read() continues reading from that position.

## 8.4 Pickles

In Python, a pickle refers to a serialized representation of a Python object, which allows you to save objects to a file or transmit them over a network and later load them back into memory in their original form.

The process of converting a Python object into a byte stream is called pickling, and the reverse process is called unpickling.


Key points:

- Can serialize most Python objects (lists, dictionaries, classes, etc.).
- Not all objects are pickleable (e.g., open file handles, sockets).
- Pickled files are binary, not human-readable.


We use the `dump()` function, which is inside the pickle module to crete a picke. So at the beginning of your program you will have to write `import pickle`. Then open an empty file, and use `pickle.dump()` to drop the object into that file.

# 9. Exception Handling

Exception handling in Python is a mechanism to handle runtime errors (exceptions) so that the program can continue execution instead of crashing.

## 9.1 Errors


In Python:

- An Error is: A problem in the program that makes it unable to run.
- Usually caused by syntax mistakes or invalid code structure.

Example:

`print("Hello" `  # SyntaxError: unexpected EOF while parsing

## 9.2 Exceptions
- Exception: An error that occurs while the program is running (runtime).
- A exception can often be handled using try–except blocks.

Example:

`x = 10 / 0  ` # ZeroDivisionError (an exception)

Syntax for handling exceptions in Python using try-except block.

```
try:
    # code that may raise an exception
except SomeException as e:
    # code to handle the exception
else:
    # code to execute if no exception occurs
finally:
    # code that always executes

```
Explanation:

- try → code that might fail
- except → handles specific errors
- else → runs if no exception occurs
- finally → always runs (cleanup, closing files, etc.)


## 9.2.1 list of common Python exception

- ZeroDivisionError → when dividing by zero.

- ValueError → when a function gets the right type of argument but with an invalid value.

- TypeError → when an operation is applied to an object of the wrong type.

- IndexError → when accessing an invalid list/tuple index.

- KeyError → when accessing a dictionary with a missing key.

- NameError → when a variable/function name is not defined.

- FileNotFoundError → when trying to open a file that doesn’t exist.

- AttributeError → when trying to access an attribute/method that an object doesn’t have.

- ImportError → when a module can’t be imported.

- IndentationError → when indentation is incorrect in the code.

- MemoryError → when the system runs out of memory.
